```python
# --- Part One ---

import re

test = "xmul(2,4)%&mul[3,7]!@^do_not_mul(5,5)+mul(32,64]then(mul(11,8)mul(8,5))"

pattern = r"mul\((\d{1,3}),(\d{1,3})\)"

products = []

with open("input.txt","r") as file:
    for line in file:

        matches = re.findall(pattern, line)
        #print(matches)

        for match in matches:
            n1, n2 = map(int, match)
            products.append(n1*n2)
            #print(n1,n2, n1*n2)

print (f"The solution of Puzzle 1 is: {sum(products)}")
```

```python
# --- Part Two ---

import re

test = "xmul(2,4)&mul[3,7]!^don't()_mul(5,5)+mul(32,64](mul(11,8)undo()?mul(8,5))"

mul = r"mul\((\d{1,3}),(\d{1,3})\)"
do = r"do\(\)"
dont = r"don't\(\)"
do_dont = r"(do\(\)|don't\(\))"

products = []

with open("input.txt","r") as file:
    
    # concatenate all lines in a single string
    line = ''.join(line.strip() for line in file)
    #line = test # uncomment dor using test
    print(line)

    #start=  re.search(do, line)
    #stop = re.search(dont,line)
    #print(start.end(), stop.start())
    matches = re.findall(mul, line)

    mul_matches = [(m.group(), m.start()) for m in re.finditer(mul, line)]
    dodont_matches = [(m.group(), m.start()) for m in re.finditer(do_dont, line)]


    print(mul_matches)
    print(dodont_matches)

    for mul, mul_pos in mul_matches:
        last_preceding = None
        for marker, marker_pos in dodont_matches:
            if marker_pos < mul_pos:
                last_preceding = marker
            else: 
                break
        # process if the last marker is "do()"
        if last_preceding != "don't()":
            numbers = re.findall(r"\d{1,3}", mul)
            n1, n2 = map(int, numbers)
            products.append(n1*n2)


print (f"The solution of Puzzle 2 is: {sum(products)}")
```
