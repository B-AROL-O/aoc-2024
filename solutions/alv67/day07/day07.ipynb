```python
# --- Part Two ---

from itertools import product
from icecream import ic

test = """190: 10 19
3267: 81 40 27
83: 17 5
156: 15 6
7290: 6 8 6 15
161011: 16 10 13
192: 17 8 14
21037: 9 7 18 13
292: 11 6 16 20
"""

retvalue = 0

def eval_expression(numbers, ops):
    """
    Evaluate an expression with numbers and operators left-to-right
    """
    result = numbers[0]
    for i, op in enumerate(ops):
        if op == "|":
            result = int(str(result)+str(numbers[i+1]))
        else:
            result = eval(f"{result}{op}{numbers[i+1]}")
    return result


def is_equation_possible(test_value, numbers):
    """
    check if the equation is possible
    : added | for concatenate operation
    """
    n_oper = len(numbers)-1
    for ops in product("+*|", repeat=n_oper):
        
        if eval_expression(numbers, ops) == test_value:
            return True
    return False


with open("input.txt","r") as file:
    #file = test.splitlines() # decomment to use test
    equations = [(int(a), list(map(int,b.strip().split()))) for a, b in [line.strip().split(":") for line in file]]
    for test_value, numbers in equations:
        if is_equation_possible(test_value, numbers):
            ic(test_value, numbers)
            retvalue += test_value

print (f"The solution of Puzzle 2 is: {retvalue}")

```
