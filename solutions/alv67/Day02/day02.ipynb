```python
# --- Part One ---
retvalue = 0

test = ["7 6 4 2 1","1 2 7 8 9","9 7 6 2 1","1 3 2 4 5","8 6 4 4 1","1 3 6 7 9"]

def sign(i:int):
    return -1 if i < 0 else 1

def isSafe(lt: list):
    l = lt.copy()
    a = int(l.pop(0))
    s = sign(a - int(l[0]))
    #print(lt, s)
    while l:
        b = int(l.pop(0))
        if abs(a -b) > 3 or abs(a-b) <1 or s != sign(a-b):
            #print (f"NOT SAFE: {a}, {b}, {s}")
            return False
        a = b
    return True

with open("input.txt","r") as file:
    for line in file:
        c = line.split()
        safe = isSafe(c)
        if safe:
            retvalue += 1
        # print (c, safe)

print (f"The solution of Puzzle 1 is: {retvalue}")
```

```python
# --- Part Two ---
retvalue = 0

test = ["7 6 4 2 1","1 2 7 8 9","9 7 6 2 1","1 3 2 4 5","8 6 4 4 1","1 3 6 7 9"]

def sign(i:int):
    return -1 if i < 0 else 1

def rm_element(l: list, n: int) -> list:
    if n == 0:
        return l[1:]
    elif n == len(l) - 1:
        return l[:-1]
    else:
        return l[:n] + l[n+1:]

def isSafe(lt: list, newtry: bool = False) -> bool:
    l = [int(_) for _ in lt]
    ckval = [int(1<= abs(a-b) <=3) for a, b in zip(l, l[1:])]
    cksign =  [int(sign(l[0]-l[1]) == sign(a-b)) for a, b in zip(l, l[1:])]
    # print (cksign)
    # se il signo è diverso da tutti quelli successivi allora elimino
    # il primo valore e testo l'array [1:]
    if len(cksign) < 2:
        print(f"                        !!!! {lt} !!!!")
    if cksign[0] and not any(cksign[1:]):
        cksign = [int(not _) for _ in cksign]
        print(f"--->{cksign}")
    checkall = [a & b for a, b in zip(ckval, cksign)]
    print (checkall)
    if not all(checkall):
        if newtry: # can try only one time
            return(False)
        else:
            idx = checkall.index(0)
            #print(idx)
            l2 = rm_element(lt, idx+1)
            if idx == 0:
                l1 = rm_element(lt, 0)
                return isSafe(l1, True) or isSafe(l2, True)
            else:
                return isSafe(l2, True)
                
    # if not all(cksign):
    #    return False
    
    return True


with open("input.txt","r") as file:
    for line in file:
        c = line.split()
        safe = isSafe(c)
        retvalue = retvalue + 1 if safe else retvalue
        print (c, safe)

print (f"The solution of Puzzle 2 is: {retvalue}")
```
