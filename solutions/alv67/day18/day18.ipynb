```python
# --- Part One and Part Two ---

memory = {"size": (71,71), "filename": "input.txt", "n_falls": 1024}

#memory = {"size": (7,7), "filename": "test.txt", "n_falls": 12} # decomment to test

class Memory:
    def __init__(self, memory: dict):
        self.size = memory["size"]
        self.memory = []
        self._init_memory()
        self.bad_blocks = []
        self.exit_path = []
        self.load_file(memory["filename"])

    def _init_memory(self):
        self.memory = [[None for _ in range(self.size[0])] for _ in range(self.size[1])]
    def load_file(self, filename: str):
        with open(filename, 'r') as file:
            #print([(int(x),int(y)) for x, y in [line.strip().split(',') for line in file]])
            self.bad_blocks = [(int(x),int(y)) for x, y in [line.strip().split(',') for line in file]]
 
    def corrupt(self, qty = None):
        self._init_memory()
        if qty: bad_blocks = self.bad_blocks[:qty]
        else: bad_blocks = self.bad_blocks 
        for x, y in bad_blocks:
            self.memory[y][x] = "#"
    
    def add_corrupt(self, pos):
        x,y = pos
        self.memory[y][x] = "#"

    def find_path(self,start = (0,0)):
        directions = [(1,0), (0,1), (-1,0), (0, -1)]
        map_cost= {}
        pos = start
        end = (self.size[0]-1,self.size[1]-1)
        map_cost[pos] = 0
        to_check = [pos]
        while to_check:
            p = to_check.pop(0)
            c = map_cost[p]
            for d in directions:
                x, y = (p[0]+d[0], p[1]+d[1])
                if x < 0 or x >= self.size[0] or y < 0 or y >= self.size[1]:
                    continue
                if self.memory[y][x]: 
                    continue
                if (x, y) not in map_cost or map_cost[(x,y)] > c+1:
                    map_cost[(x,y)] = c+1
                    to_check.append((x,y))

        ret = map_cost.get(end, None)
        
        # get exit path
        # exit_path = []
        # to_check = [end]
        # while ret and to_check:
        #     p = to_check.pop(0)
        #     c = map_cost[p]
        #     for d in directions:
        #         x, y = (p[0]+d[0], p[1]+d[1])
        #         if x < 0 or x >= self.size[0] or y < 0 or y >= self.size[1]:
        #             continue
        #         if (x, y) in map_cost and map_cost[(x,y)] == c-1:
        #             exit_path.append((x,y))
        #             to_check.append((x,y))

        # self.exit_path = exit_path
        return ret

    def __str__(self):
        retstr = ""
        for y, row in enumerate(self.memory):
            for x, cell in enumerate(row):
                if cell == "#":
                    retstr += "#"
                elif (x,y) in self.exit_path:
                    retstr += "O"
                else:
                    retstr += "."
            retstr += "\n"
        return retstr

# ---- main ------

mem_space = Memory(memory)

mem_space.corrupt(memory["n_falls"])
# print(mem_space)
res1 = mem_space.find_path()
print(f"The solution for part 1 is: {res1}")


# part 2
res2 = 0
bad_blocks = mem_space.bad_blocks
bb_ok = memory["n_falls"]
bb_max = len(bad_blocks)
bb_test = 0
while bb_max - bb_ok > 1:
    bb_test = bb_ok + (bb_max-bb_ok)//2
    #print(f"Test with {bb_test} bad blocks")
    mem_space.corrupt(bb_test)
    ret = mem_space.find_path()
    if ret:
        #print(f"  Path Found")
        bb_ok = bb_test
    else:
        #print(f"  !! no Path !!")
        bb_max = bb_test
#print (f">>> {bb_ok} <<<< {bad_blocks[bb_ok]}")
res2 = bad_blocks[bb_ok]

print(f"The solution for part 2 is: {res2}")

```
