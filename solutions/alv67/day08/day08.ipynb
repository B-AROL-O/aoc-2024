```python
# --- Part One ---

from itertools import combinations
from icecream import ic

test = """............
........0...
.....0......
.......0....
....0.......
......A.....
............
............
........A...
.........A..
............
............
"""

retvalue = 0

class Roof:
    """
        antennas e un dizionario con chiave tipo di antenna e valore lista (x,y) delle posizioni delle antenne
    """
    def __init__(self, grid):
        self.grid =[list(c) for c in [l.strip() for l in grid]]
        self.size = (len(self.grid[0]), len(self.grid))
        self.antennas = self.map_antennas(self.grid)
        self.antinodes = self.map_antinodes(self.antennas)
        # ic(self.antinodes)
    
    def _check_point(self,point):
        return point[0] >= 0 and point[0] < self.size[0] and point[1] >= 0 and point[1] < self.size[1]

    def map_antennas(self, grid):
        a = {}
        for y, row in enumerate(grid):
            for x, c in enumerate(row):
                if c != ".":
                    a.setdefault(c, []).append((x, y))
        return a    

    def map_antinodes(self, ant):
        anodes = set()
        for k, v in ant.items():
            for c in combinations(v, 2):
                dx = c[0][0] - c[1][0]
                dy = c[0][1] - c[1][1]
                a1 = (c[0][0] + dx, c[0][1] + dy)
                if self._check_point(a1): anodes.add(a1)
                a2 = (c[1][0] - dx, c[1][1] - dy)
                if self._check_point(a2): anodes.add(a2)
        return anodes


with open("input.txt","r") as file:
    # file = test.splitlines() # decomment to use test
    roof = Roof(file)
    ic(roof.antinodes)
    retvalue = len(roof.antinodes)

print (f"The solution of Puzzle 1 is: {retvalue}")
```

```python
# --- Part Two ---

# This solution uses objects defined in part 1

retvalue = 0

class Roof_new(Roof):

    def map_antinodes(self, ant):
        anodes = set()
        for k, v in ant.items():
            for c in combinations(v, 2):
                delta = (c[0][0] - c[1][0], c[0][1] - c[1][1])
                an = c[0]
                while self._check_point(an):
                    anodes.add(an)
                    an = (an[0] + delta[0], an[1] + delta[1])
                an = c[1]
                while self._check_point(an):
                    anodes.add(an)
                    an = (an[0] - delta[0], an[1] - delta[1])
        return anodes

with open("input.txt","r") as file:
    # file = test.splitlines() # decomment to use test
    roof = Roof_new(file)
    #ic(roof.antinodes)
    retvalue = len(roof.antinodes)

print (f"The solution of Puzzle 2 is: {retvalue}")

```
