```python
# --- Part One and Part Two ---

from icecream import ic

test = """RRRRIICCFF
RRRRIICCCF
VVRRRCCFFF
VVRCCCJFFF
VVVVCJJCFE
VVIVCCJJEE
VVIIICJJEE
MIIIIIJJEE
MIIISIJEEE
MMMISSJEEE
"""

test = """AAAAAA
AAABBA
AAABBA
ABBAAA
ABBAAA
AAAAAA
"""

def fence_sides(fence_map: set) -> int:
    retvalue = 0
    fences = sorted(list(fence_map))
    last_d = None
    last_a = None
    last_b = None
    for d, a, b in fences:
        if d == last_d and a == last_a:
            if b != last_b + 1:
                retvalue += 1
        else:
            retvalue += 1
        last_d = d
        last_a = a
        last_b = b
    return retvalue

def find_region(grid:list, x, y, mapped: dict):
    fence = 0
    area = 0
    fence_map = set() # ("hu"|"hd"|"vl"|"vr", a, b)
    w = len(grid[0])
    h = len(grid)
    plant = grid[y][x]
    to_check = [(x, y)] # list of coordinates to check

    while to_check:
        x, y = to_check.pop()
        if grid[y][x] != plant: continue
        if (x, y) in mapped: continue
        mapped[(x, y)] = plant
        area += 1
        if x == 0:
            fence += 1 # border fence
            fence_map.add(("vl", x, y))
        else: # check left
            if grid[y][x-1] != plant:
                fence += 1
                fence_map.add(("vl", x, y))
            elif (x-1, y) not in to_check:
                to_check.append((x-1, y))

        if x == w-1:
            fence += 1 # border fence
            fence_map.add(("vr", x+1, y))
        else: # check right
            if grid[y][x+1] != plant:
                fence += 1
                fence_map.add(("vr", x+1, y))
            elif (x+1, y) not in to_check:
                to_check.append((x+1, y))

        if y == 0:
            fence += 1 # border fence
            fence_map.add(("hu", y, x))
        else: # check up
            if grid[y-1][x] != plant:
                fence += 1
                fence_map.add(("hu", y, x))
            elif (x, y-1) not in to_check:
                to_check.append((x, y-1))

        if y == h-1:
            fence += 1 # border fence
            fence_map.add(("hd", y+1, x))
        else: # check down
            if grid[y+1][x] != plant:
                fence += 1
                fence_map.add(("hd", y+1, x))
            elif (x, y+1) not in to_check:
                to_check.append((x, y+1))

    # ic (plant, fence_map)
    sides = fence_sides(fence_map)

    return(fence, area, sides)


def map_regions(grid: list):
    mapped = dict()
    regions = {} # id: [auto] , {plant: [string], fences: [int], area: [int]}
    regions_id = 0 
    for y, row in enumerate(grid):
        for x, _ in enumerate(row):
            if (x, y) in mapped: continue 
            fence, area, sides = find_region(grid, x, y, mapped)
            regions[regions_id] = {"plant": grid[y][x], "fences": fence, "area": area, "sides": sides}
            regions_id += 1
    return regions

with open("input.txt","r") as file:
    # file = test.splitlines() # decomment to use test
    garden =  [list(c) for c in [line.strip() for line in file]]

    regions = map_regions(garden)

    #ic(regions)

    ret1 = sum(r["fences"] * r["area"] for r in regions.values())
    ret2 = sum(r["sides"] * r["area"] for r in regions.values())


print (f"The solution of for part 1 is: {ret1}")
print (f"The solution of for part 2 is: {ret2}")

```
